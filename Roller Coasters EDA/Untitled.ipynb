## Understanding the Features (Univariate Analysis)

let's explore the year introduced column

```python
df['Year_Introduced'].value_counts()
```

In the year 1999, there were 46 roller coasters introduced. The value counts shows the frequency of the year from most to least

```python
df['Year_Introduced'].value_counts().head(10).plot(kind = 'bar', title = 'Top Years Coasters Introduced')
```

adding axes names:

```python
ax = df['Year_Introduced'].value_counts().head(10).plot(kind = 'bar', title = 'Top Years Coasters Introduced')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
```
