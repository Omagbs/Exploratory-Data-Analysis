### what if we wanted to compare more than two features?

```python
sns.pairplot(df, vars = ['Year_Introduced', 'Speed_mph','Height_ft', 'Inversions', 'Gforce'])
```

Converting infinity values to NaN

## Conlcusion: Asking a Question about the Data
Q. What the locations with the fastest Roller Coasters (Minimum of 10 a that location)

```python
df['Location'].value_counts()
```

```python
ax = df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean', 'count']).query('count >= 10').sort_values('mean')['mean'].plot(kind='barh', figsize = (12,5), title = 'Average Coaster Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
